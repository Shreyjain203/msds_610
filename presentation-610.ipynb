```python
# Improt libraries
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# Read 'Credit.csv'
df = pd.read_csv("Credit.csv")
credit = df[['Income', 'Limit', 
            'Rating', 'Age', 'Balance']]
credit.info()
```

```python
# Print first 5 obesrvation
credit.head()
```

```python
model =smf.ols('Balance~Income + Limit +Rating + Age',data=credit).fit()
model.summary()
```
